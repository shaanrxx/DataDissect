```python
# import libraries
import requests
from bs4 import BeautifulSoup

import functools as ft
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
```

```python


r = requests.get("https://logicalharmony.net/cruelty-free-brand-list-brands-to-avoid/")
soup = BeautifulSoup(r.text, 'html.parser')

# create an empty list that will store brands
brands = []

# create another list that will contain each list item still with the tags
rows = []

# get all list tags in entry-content part of the page (found by inspecting site)
brand_section1 = soup.find(class_='entry-content')
for brand in brand_section1.find_all('li'):
    # make sure brands are being extracted correctly
    # print(brand.text)
    # append each brand to the list of brands
    brands.append(brand.text)
    # keep the tags info in this list
    rows.append(brand)
    
len(brands)
```

```python
results = []
for brand in brands:
    # search for the brand on google
    query = f"{brand} sephora"
    search_url = f"https://www.google.com/search?q={query}"
    r = requests.get(search_url)
    soup = BeautifulSoup(r.text, 'html.parser')

    # check if Sephora's website is among the search results
    for result in soup.find_all('cite'):
        if 'sephora' in result.text:
            print(f"{brand} is available on Sephora")
            break
    else:
        print(f"{brand} is not available on Sephora")
```

```python
# create a pandas dataframe with the results
df = pd.DataFrame(results, columns=['Brand', 'Available on Sephora'])
print(df.head())
```

```python
data = []

for brand in brands:
    # search for the brand on google
    query = f"{brand} sephora"
    search_url = f"https://www.google.com/search?q={query}"
    r = requests.get(search_url)
    soup = BeautifulSoup(r.text, 'html.parser')

    # check if Sephora's website is among the search results
    available_on_sephora = False
    for result in soup.find_all('cite'):
        if 'sephora' in result.text:
            available_on_sephora = True
            break

# create a dictionary to store the brand name and availability info
    brand_dict = {'brand_name': brand, 'available_on_sephora': available_on_sephora}
    data.append(brand_dict)
```

```python

# convert the list of dictionaries into a pandas DataFrame
df = pd.DataFrame(data)
df.to_csv('C:/Users/shaan/Desktop/big_data/bsc-big-data/Project/DataSets/NCFB_available.csv', index=False)
```
