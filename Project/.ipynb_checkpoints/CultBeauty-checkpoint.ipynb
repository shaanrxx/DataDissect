```python
# import libraries
import requests
from bs4 import BeautifulSoup

import matplotlib.pyplot as plt
import functools as ft
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import re
    
```

```python

r = requests.get("https://logicalharmony.net/cruelty-free-brands-at-cult-beauty/")
soup = BeautifulSoup(r.text, 'html.parser')

# create an empty list that will store brands
brands = []

# create another list that will contain each list item still with the tags
rows = []

# get all list tags in entry-content part of the page (found by inspecting site)
brand_section1 = soup.find(class_='entry-content')
for brand in brand_section1.find_all('li'):
    # make sure brands are being extracted correctly
    # print(brand.text)
    # append each brand to the list of brands
    brands.append(brand.text)
    # keep the tags info in this list
    rows.append(brand)
    
len(brands)
```

```python
# create dataframe - start with just brands list, then adjust
cruelty_free_cultbeauty_df = pd.DataFrame({'Brand_info': brands})   
cruelty_free_cultbeauty_df.to_csv('C:/Users/shaan/Desktop/big_data/bsc-big-data/Project/DataSets/cruelty_free_cultbeauty.csv', index=False)
cruelty_free_cultbeauty_df.head()
```

```python
# Send a GET request to the website and retrieve the HTML content
url = "https://www.cultbeauty.com/brands.list"
response = requests.get(url)
html_content = response.text

# Create a BeautifulSoup object to parse the HTML content
soup = BeautifulSoup(html_content, "html.parser")

# Find all the 'a' tags within the HTML content that contain the brand names
brand_tags = soup.find_all("a", attrs={"class": "responsiveBrandsPageScroll_brand"})

# Extract the brand names from the 'a' tags
brands = [tag.text.strip() for tag in brand_tags]

# Print the brand names
print(brands)


# Create a DataFrame from the extracted data
data = {"Brand_info": brands}
all_brands_cultbeauty_df = pd.DataFrame(data)

# Save the DataFrame as a CSV file
all_brands_cultbeauty_df.to_csv("C:/Users/shaan/Desktop/big_data/bsc-big-data/Project/DataSets/all_brands_cultbeauty.csv", index=False)

# Print the DataFrame
print(all_brands_cultbeauty_df)
```

```python
# Load data from CSV files
all_brands_cultbeauty_df = pd.read_csv('C:/Users/shaan/Desktop/big_data/bsc-big-data/Project/DataSets/all_brands_cultbeauty.csv')
cruelty_free_cultbeauty_df = pd.read_csv('C:/Users/shaan/Desktop/big_data/bsc-big-data/Project/DataSets/cruelty_free_cultbeauty.csv')

# Extract brand names from each DataFrame and clean them
all_brands = all_brands_cultbeauty_df['Brand_info'].str.strip().str.lower().tolist()
cruelty_free_brands = cruelty_free_cultbeauty_df['Brand_info'].str.strip().str.lower().tolist()

# Create a DataFrame to store the results
results_df = pd.DataFrame({'Brand_info': all_brands})
results_df['cruelty_free'] = results_df['Brand_info'].str.lower().isin(cruelty_free_brands)

# Print the results
print(results_df.head(40))
```

```python
# Count the number of cruelty-free brands
num_cruelty_free_brands = results_df['cruelty_free'].sum()

# Print the number of cruelty-free brands
print("Number of cruelty-free brands:", num_cruelty_free_brands)
```

```python
# Convert brand names to lowercase for case-insensitive comparison
all_brands_cultbeauty_df['Brand_info'] = all_brands_cultbeauty_df['Brand_info'].str.lower()
cruelty_free_cultbeauty_df['Brand_info'] = cruelty_free_cultbeauty_df['Brand_info'].str.lower()

# Check if brand names appear in both data frames
all_brands_cultbeauty_df['cruelty_free'] = all_brands_cultbeauty_df['Brand_info'].isin(cruelty_free_cultbeauty_df['Brand_info'])

# Calculate the number of cruelty-free brands out of all brands
num_cruelty_free_brands = all_brands_cultbeauty_df['cruelty_free'].sum()

# Calculate the percentage of cruelty-free brands out of all brands
percentage_cruelty_free = (num_cruelty_free_brands / len(all_brands_cultbeauty_df)) * 100

# Calculate the percentage of non-cruelty-free brands out of all brands
percentage_non_cruelty_free = 100 - percentage_cruelty_free

# Print the results
print(f"Number of cruelty-free brands out of all brands: {num_cruelty_free_brands}")
print(f"Percentage of cruelty-free brands out of all brands: {percentage_cruelty_free:.2f}%")
print(f"Percentage of non-cruelty-free brands out of all brands: {percentage_non_cruelty_free:.2f}%")
```

```python
# Calculate the number of non-cruelty-free brands
num_non_cruelty_free_brands = len(all_brands_cultbeauty_df) - num_cruelty_free_brands

# Create a list of labels for the pie chart
labels = ['Cruelty-Free', 'Non-Cruelty-Free']

# Create a list of values for the pie chart
values = [num_cruelty_free_brands, num_non_cruelty_free_brands]

# Create a pie chart
plt.pie(values, labels=labels, autopct='%1.1f%%')

# Add a title to the pie chart
plt.title('Cruelty-Free vs Non-Cruelty-Free Brands')

# Display the pie chart
plt.show()
```

```python
# Define colors for the pie chart slices
colors = ['#55a868', '#c44e52']

# Define explode values to emphasize a particular slice (optional)
explode = [0.1, 0]

# Create a pie chart
plt.pie(values, labels=labels, explode=explode, colors=colors, autopct='%1.1f%%', startangle=90, shadow=True)

# Add a title and legend to the pie chart
plt.title('Cruelty-Free Brands verified by Logic Harmony in Cult Beauty')
plt.legend(loc='upper right', labels=labels)

# Set aspect ratio to make the pie chart a circle
plt.axis('equal')

# Display the pie chart
plt.show()
```

```python
import pandas as pd

# Load data from CSV files
all_peta_df = pd.read_csv('C:/Users/shaan/Desktop/big_data/bsc-big-data/Project/DataSets/all_peta.csv')
all_brands_cultbeauty_df = pd.read_csv('C:/Users/shaan/Desktop/big_data/bsc-big-data/Project/DataSets/all_brands_cultbeauty.csv')

# Convert brand names to lowercase for case-insensitive comparison
all_peta_df['Brand_info'] = all_peta_df['Brand_info'].str.lower()
all_brands_cultbeauty_df['Brand_info'] = all_brands_cultbeauty_df['Brand_info'].str.lower()

# Add a column to store the result of cruelty-free check
all_brands_cultbeauty_df['cruelty_free'] = all_brands_cultbeauty_df['Brand_info'].isin(all_peta_df['Brand_info'])

# Calculate the number of cruelty-free products
num_cruelty_free = all_brands_cultbeauty_df['cruelty_free'].sum()

# Calculate the percentage of cruelty-free products out of all products
percentage_cruelty_free = (num_cruelty_free / len(all_brands_cultbeauty_df)) * 100

# Calculate the percentage of non-cruelty-free products out of all products
percentage_non_cruelty_free = 100 - percentage_cruelty_free

# Print the updated DataFrame
print(all_brands_cultbeauty_df.head(50))

# Print the number of cruelty-free products and their percentages
print(f"Number of cruelty-free products: {num_cruelty_free}")
print(f"Percentage of cruelty-free products: {percentage_cruelty_free:.2f}%")
print(f"Percentage of non-cruelty-free products: {percentage_non_cruelty_free:.2f}%")
```

```python
import matplotlib.pyplot as plt

# Calculate the number of cruelty-free and non-cruelty-free products
num_cruelty_free = all_brands_cultbeauty_df['cruelty_free'].sum()
num_non_cruelty_free = len(all_brands_cultbeauty_df) - num_cruelty_free

# Create labels for the pie chart
labels = ['Cruelty-Free', 'Non-Cruelty-Free']

# Create colors for the pie chart
colors = ['#57B89E', '#E37222']

# Create the pie chart
plt.pie([num_cruelty_free, num_non_cruelty_free], labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)

# Add a title
plt.title('Cruelty-Free Products in Cult Beauty Verified my PETA')

# Add a legend
plt.legend()

# Display the chart
plt.show()
```
