```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.gridspec import GridSpec

# Define the number of retailers
num_retailers = 4

# Define the position of each bar on the x-axis
bar_width = 0.35
index = np.arange(num_retailers)

# Create a bar graph for each retailer
def create_bar_graph(ax, retailer_name, cruelty_free_values, non_cruelty_free_values):
    # Create the bars for cruelty-free brands
    rects1 = ax.bar(index, cruelty_free_values, bar_width, label='Cruelty-Free', color='#009BDB')

    # Create the bars for non-cruelty-free brands
    rects2 = ax.bar(index + bar_width, non_cruelty_free_values, bar_width, label='Non-Cruelty-Free', color='#FF1215')

    # Add data labels on top of each bar
    def add_labels(rects):
        for rect in rects:
            height = rect.get_height()
            ax.annotate('{}'.format(height),
                        xy=(rect.get_x() + rect.get_width() / 2, height),
                        xytext=(0, 3),
                        textcoords="offset points",
                        ha='center', va='bottom')

    add_labels(rects1)
    add_labels(rects2)

    # Customize the plot
    ax.set_xlabel('Retailer')
    ax.set_ylabel('Number of Brands')
    ax.set_title(retailer_name)
    ax.set_xticks(index + bar_width / 2)
    ax.set_xticklabels(labels)
    ax.legend()

# Create subplots for each retailer
fig = plt.figure(figsize=(12, 9))
gs = GridSpec(2, 2)

# Retailer 1: Beauty Bay
ax1 = fig.add_subplot(gs[0, 0])
create_bar_graph(ax1, 'Beauty Bay', values_cruelty_free, values_non_cruelty_free)

# Retailer 2: Cult Beauty
ax2 = fig.add_subplot(gs[0, 1])
create_bar_graph(ax2, 'Cult Beauty', values_cruelty_free, values_non_cruelty_free)

# Retailer 3: ULTA
ax3 = fig.add_subplot(gs[1, 0])
create_bar_graph(ax3, 'ULTA', values_cruelty_free, values_non_cruelty_free)

# Retailer 4: Sephora
ax4 = fig.add_subplot(gs[1, 1])
create_bar_graph(ax4, 'Sephora', values_cruelty_free, values_non_cruelty_free)

# Adjust the layout
fig.tight_layout()

# Display the plot
plt.show()
```
